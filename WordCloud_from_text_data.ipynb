# WordCloud from Text Data

```python
#Importando bibliotecas
import re
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from os import path
import matplotlib.pyplot as plt
```

```python
#Lendo o arquivo
text = open('CV_3.txt',"r")
data = text.read() 
#data
```

```python
#Definindo a lista de stopwords
stopwords= set(STOPWORDS)

#Adicionando a lista stopwords em portuguÃªs
new_words = []
with open("stopwords_portuguese.txt", 'r') as f:
    [new_words.append(word) for line in f for word in line.split()]

new_stopwords = stopwords.union(new_words)
#new_stopwords
```

```python
%matplotlib inline
```

```python
#Printando a wordcloud
plt.figure(figsize=(20,10))
wc = WordCloud(min_font_size=10, 
               max_font_size=300, 
               background_color='white', 
               mode="RGB",
               stopwords=new_stopwords,
               width=2000, 
               height=1000,
               normalize_plurals= True).generate(data)


plt.title("CV Wordcloud", fontsize=40, color="red")
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
